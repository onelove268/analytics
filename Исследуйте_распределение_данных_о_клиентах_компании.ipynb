### Подсказки:

* Проверьте струкуру данных и тиры переменных с помощью функции head() и dtypes().
* Для обнаружения пропущенных значени используйте функцию isnull().sum().
* Используйте функцию describe() для получения описательныз статистик данных.
* Для построениея гистограммы и ящика с усами используйте библиотеку seaborn.
* Для проверки нормальности распределения используйте функцию shapiro() из модуля scipy.stats.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro

# Загрузка данных из файла CSV
data = pd.read_csv('/content/data_test (1).csv')

# Предварительный анализ данных
# Изучение структуры данных
print('Структура данных:')
print(data.head())
print()

# Оценка пропущенных значений
print('Пропущенные значения:')
print(data.isnull())
print()

# Проверка аномальных значений и выбросов
print('Описание данных:')
print(data.describe())
print()

# Исследование распределения данных
# Построение гистораммы
plt.figure(figsize=(8, 6))
sns.histplot(data['variable'], kde=True)
plt.title('Распределение данных')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.show()

# Построение ящика с усами
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['variable'])
plt.title('Ящик с усами')
plt.xlabel('Переменная')
plt.show()

# Проверка нормальности распределения
stat, p = shapiro(data['variable'])
print('Тест нормальности:')
print(f'Статистика теста: {stat}')
print(f'p-value: {p}')
print()
```

### Выводы анализа:

По результатм анализа распределения переменной "variable" можно сделать следующие выводы:

1. Гистограмма показвает, что распределение данных имеет "хвост".
2. Описательные статистики:
* среднее значение 21.1
* стандартное отклонение 9.6
3. Ящик с усами показывает наличие выбросов в районе 60
4. Посколько p-value меньше 0.05 мы отвергаем нулевую гипотезу теста Шапиро-Уилка. Это означет, что у нас есть достаточно доказательств, чтобы утверждать, что выборки не получены из нормального распределенния.
